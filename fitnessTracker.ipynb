```python
# Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#1 - Data generation
# Defining parameters for each cluster
np.random.seed(42)

#Number of users for each cluster (random interval of 100 between 100 and 1000)
N1 = np.random.randint(100, 1000)
N2 = np.random.randint(100, 1000)
N3 = np.random.randint(100, 1000)

#Cluster 1 - Active)
st1 = np.random.randint(8000, 12001, N1)
sl1 = np.random.uniform(7, 9, N1)

#Cluster 2 - Moderately active
st2 = np.random.randint(5000, 8001, N2)
sl2 = np.random.uniform(6, 7.5, N2)

#Cluster 3 - Least active
st3 = np.random.randint(2000, 5001, N3)
sl3 = np.random.uniform(5, 6.5, N3)

#Combining the data
steps = np.concatenate([st1, st2, st3])
sleep = np.concatenate([sl1, sl2, sl3])

# Creating the feature matrix
X = np.column_stack((steps, sleep))





#2 - Visualizing the Data
plt.scatter(X[:, 0], X[:, 1], c='gray', alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel("Steps per Day")
plt.ylabel("Hours Slept per Night")
plt.title("Fitness Tracker Data Visualization")
plt.show()




#3 - KMeans Clustering
# Defining the number of clusters
nClusters = 3

# Creating a KMeans object with the specified number of clusters
kMeans = KMeans(n_clusters=nClusters, random_state=42)

# Fitting the KMeans model to the data
kMeans.fit(X)

# Getting the cluster labels assigned by KMeans
kMeansLabels = kMeans.labels_







#4 - Cluster Visualization
plt.scatter(X[:, 0], X[:, 1], c=kMeansLabels, cmap='viridis', alpha=0.6, edgecolors='w', linewidth=0.5)
plt.xlabel("Steps per Day")
plt.ylabel("Hours Slept per Night")
plt.title("KMeans Clustering of Fitness Tracker Data")
plt.show()






#5 - Cluster Analysis
# Calculating the average steps and sleep values for each cluster
for cluster in range(nClusters):
    mask = (kMeansLabels == cluster)
    aveSt = np.mean(X[mask, 0])
    aveSl = np.mean(X[mask, 1])
    print(f"Cluster {cluster + 1}: Average Steps = {aveSt:.2f}, Average Sleep = {aveSl:.2f} hours")






```
